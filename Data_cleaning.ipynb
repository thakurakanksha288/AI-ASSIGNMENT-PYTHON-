```python
import numpy as np
import pandas as pd

# Load the dataset from the specified path
path="/content/students_performance_raw.csv"
df=pd.read_csv(path)

# Display the first few rows of the raw DataFrame
df.head()

# Calculate the median age to use for filling missing age values
median=df['age'].median()
print(f"Median age: {median}")

# --- Data Cleaning and Preprocessing Steps ---

# Fill NaN values in the 'age' column with the calculated median
df['age'] = df['age'].fillna(median)

# Handle missing and invalid values in score columns ('maths_score', 'science_score', 'english_score')
score_columns = ['maths_score', 'science_score', 'english_score']
for col in score_columns:
    # Replace '-1' entries with NaN, as -1 is likely an indicator of missing data
    df[col] = df[col].replace(-1, np.nan)
    # Fill any remaining NaN values (original NaNs and those from -1) with the mean of the column
    df[col] = df[col].fillna(df[col].mean())
    # Enforce score constraints: not negative and not above 100
    df[col] = df[col].clip(lower=0, upper=100)

# Standardize the 'gender' column:
# 1. Replace 'F' with 'female' and 'Male' with 'male'
# 2. Convert all entries to lowercase for consistency
df['gender'] = df['gender'].replace({'F': 'female', 'Male': 'male'}).str.lower()

# Fill NaN values in the 'name' column with 'Unknow'
df['name'] = df['name'].fillna('Unknow')

# Standardize the 'placement_interest' column:
# 1. Replace 'YES', 'Y' with 'yes'
# 2. Replace 'NO', 'N', 'Maybe' with 'no'
# 3. Convert all entries to lowercase for consistency
df['placement_interest'] = df['placement_interest'].replace({'YES': 'yes', 'NO': 'no','Maybe':'no','Y':'yes','N':'no'}).str.lower()

# Correct age values that are 100 or greater by replacing them with the median age
# This assumes ages 100+ are data entry errors
df.loc[df['age'] >= 100, 'age'] = df['age'].median()

# Handle the 'attendance' column:
# 1. Remove the '%' character
# 2. Convert the column to a float data type
# 3. Calculate the mean attendance from the cleaned numeric values
# 4. Fill any NaN values with the calculated mean attendance
# 5. Round the attendance values and convert to integer type
df['attendance'] = df['attendance'].str.replace('%', '', regex=False).astype(float)
mean_attendance = df['attendance'].mean()
df['attendance'] = df['attendance'].fillna(mean_attendance)
df['attendance'] = df['attendance'].round().astype(int)

# Display the first few rows of the cleaned DataFrame
df.head()
```
