```python
##numpy library##


#making an array
import numpy as np
array=np.array([1,2,3,5,4])
print(array)
print(type(array))


#2d array
arr=np.array([[1,2,3],[5,6,4]])
print(arr)

#3d array
a=np.array([[[1,2,3],[5,2,3]], [[5,6,9],[1,5,6]]])

#numpy function

print(a.ndim)
print(arr[1,2])
print(arr[:4])
print(arr.shape)
n=np.array([1,5,3,5])
newarr=n.reshape(2,2)
print(newarr)
newr=arr.reshape(-1)
print(newr)


#nditer() printing each item of the array
for x in np.nditer(a):
  print(x)


#for the index of the array
for idx,x in np.ndenumerate(arr):
  print(idx,x)


#for concatenating two array(same size of the array is must)
r=np.concatenate((n,array))
print(r)

#for joining we use stack

arr1=np.array([1,2,3])
arr2=np.array([4,5,6])

arr3=np.stack((arr1,arr2),axis=1)
print(arr3)

#along horizontal axis
arr4=np.hstack((arr1,arr2))
print(arr4)

#for spliting two array
arr5=np.array_split(arr,4)
print(arr5)

#for sereaching two arrays
y=np.where(arr==4)
print(y)

#for a sorted array
z=np.searchsorted(array,7)
print(z)

#for sorting an array

c=np.sort(arr)
print(c)
```

```python
#scikit-learn
#used for clustingt and model selction and also for dimensionality and pre-processing of the data too


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split


win=pd.read_csv("/content/silver_price_forecast_2026.csv")
print(win)
win.head()
win.isnull().sum()


#preprocessing data
bins=(40,50.50,57)
gourp_names=["bad","good"]
win["Predicted_Price"]=pd.cut(win["Predicted_Price"],bins=bins,labels=gourp_names)


#best 5 from the data
label_quality=LabelEncoder()
win["Predicted_Price"]=label_quality.fit_transform(win["Predicted_Price"])
win.head()

#ploting the data
win['Predicted_Price'].value_counts()
sns.countplot(x='Predicted_Price', data=win)


#now seperating the values of the dataset

X=win.drop(["Predicted_Price", "Date"],axis=1) # Dropping 'Date' column here
y=win["Predicted_Price"]

#for training AND testing of the data


X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)


#applying standard scaling to get optimized result

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)

#comparing different models


#RANDOM FOREST CLASSIFIER

rfc=RandomForestClassifier(n_estimators=200)#only value you use for this
rfc.fit(X_train,y_train)
pred_rfc=rfc.predict(X_test)

X_test[:20]

#how model performed

print(classification_report(y_test,pred_rfc))
print(confusion_matrix(y_test,pred_rfc))

#SVM CLASSIFIER

clf=svm.SVC()
clf.fit(X_train,y_train)
pred_clf=clf.predict(X_test)
#how model performed

print(classification_report(y_test,pred_rfc))
print(confusion_matrix(y_test,pred_rfc))


#NUETRAL NETWORKS
mlpc=MLPClassifier(hidden_layer_sizes=(11,11,11),max_iter=500)
mlpc.fit(X_train,y_train)
pred_mlpc=mlpc.predict(X_test)


#how model perform
print(classification_report(y_test,pred_rfc))
print(confusion_matrix(y_test,pred_rfc))
```
