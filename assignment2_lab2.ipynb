```python
##numpy library##


#making an array
import numpy as np
array=np.array([1,2,3,5,4])
print(array)
print(type(array))


#2d array
arr=np.array([[1,2,3],[5,6,4]])
print(arr)

#3d array
a=np.array([[[1,2,3],[5,2,3]], [[5,6,9],[1,5,6]]])

#numpy function

print(a.ndim)
print(arr[1,2])
print(arr[:4])
print(arr.shape)
n=np.array([1,5,3,5])
newarr=n.reshape(2,2)
print(newarr)
newr=arr.reshape(-1)
print(newr)


#nditer() printing each item of the array
for x in np.nditer(a):
  print(x)


#for the index of the array
for idx,x in np.ndenumerate(arr):
  print(idx,x)


#for concatenating two array(same size of the array is must)
r=np.concatenate((n,array))
print(r)

#for joining we use stack

arr1=np.array([1,2,3])
arr2=np.array([4,5,6])

arr3=np.stack((arr1,arr2),axis=1)
print(arr3)

#along horizontal axis
arr4=np.hstack((arr1,arr2))
print(arr4)

#for spliting two array
arr5=np.array_split(arr,4)
print(arr5)

#for sereaching two arrays
y=np.where(arr==4)
print(y)

#for a sorted array
z=np.searchsorted(array,7)
print(z)

#for sorting an array

c=np.sort(arr)
print(c)
```

```python
#USING RANDOM FROM NUMPY

import numpy as np
from numpy import random

#print any random value
x=random.randint(100)
print(x)

#for printing a float number
y=random.rand(5)
print(y)#the value is always bewtween zero and one

#for making a random array (1-d)

z=random.randint(100,size=(2))
print(z)

#for making (2-d)array
a=random.randint(100,size=(2,2))
print(a)

#for choicing a number for a certain array
b=np.array([1,2,3,4])
print(b)
c=random.choice(b)
print(c)


#for data distribution
#used in stats and data science

d=random.choice([1,2,3,4],p=[0.1,0.6,0.3,0.0],size=(100))
print(d)


#shuffle an array

print(random.shuffle(b))

```
